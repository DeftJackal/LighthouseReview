```python
#Import Modules required.
import requests as re
import os
from IPython.display import JSON
from requests.exceptions import HTTPError
```

```python
#Import api keys. These are stored as environmental variables on the computer for security. We do not want to upload these to Git.
client_id = os.environ['FourSquareClientID']
client_secret = os.environ['FourSquareClientSecret']
```

```python
# Parameters that we are searching on.
location = "Toronto,Canada"
ll = 43.6510, -79.347015
v = "20200731"

#Url for the get request. Comprised of root url, request_url, parameters and keys. 
url = "https://api.foursquare.com/v2/venues/search?near="+location+"&client_id="+client_id+"&client_secret="+client_secret+"&v="+v
```

```python
res = re.get(url)
res
```
